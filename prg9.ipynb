```python
import numpy as np, matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces as fof
from sklearn.model_selection import train_test_split as tts, cross_val_score as cvs
from sklearn.metrics import accuracy_score as acs, confusion_matrix as cm, classification_report as cr
from sklearn.naive_bayes import GaussianNB as gb

data = fof(shuffle = True, random_state = 42)
x, y = data.data, data.target
xt, xs, yt, ys = tts(x, y, test_size = 0.3, random_state = 42)
gnb = gb()
gnb.fit(xt, yt)
yp = gnb.predict(xs)
acc = acs(ys, yp)
print(f'ACS: {acc * 100:.2f}%')
print(cr(ys, yp, zero_division = 1))
print(cm(ys, yp))
```
